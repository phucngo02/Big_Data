```python
import nltk
nltk.download('punkt')
```

```python
from pyvi import ViTokenizer

# Định nghĩa hàm để tách từ và gom thành câu
def word_segmentation(text):
    sentences = nltk.sent_tokenize(text)
    output = []
    for sentence in sentences:
        segmented_sentence = ViTokenizer.tokenize(sentence)
        output.extend(segmented_sentence.split())
    return ' '.join(output)
```

```python
df['post_message'] = df['post_message'].apply(word_segmentation)
```
